```python
import pandas
import datetime
```

```python
# Read in RBSP sat A data
rbspSatADF = pandas.read_csv("../data/rbsp_iono_satA.txt", \
                             delim_whitespace=True, header=None)
rbspSatADF.columns = [ "dateStr", "timeStr", "MLatNth", "MLonNth",\
                      "MLTNth", "MLatSth", "MLonSth", "MLTSth", "sat" ]
rbspSatADF[ ["MLatNth", "MLonNth", "MLTNth", "MLatSth", "MLonSth", "MLTSth"] ] = rbspSatADF[ ["MLatNth", "MLonNth", "MLTNth",\
             "MLatSth", "MLonSth", "MLTSth"] ].astype(float)
print rbspSatADF.head()
####### DO THE SAME FOR SAT B #######
####### DO THE SAME FOR SAT B #######
rbspSatBDF = pandas.read_csv("../data/rbsp_iono_satB.txt", \
                             delim_whitespace=True, header=None)
rbspSatBDF.columns = [ "dateStr", "timeStr", "MLatNth", "MLonNth",\
                      "MLTNth", "MLatSth", "MLonSth", "MLTSth", "sat" ]
rbspSatBDF[ ["MLatNth", "MLonNth", "MLTNth", "MLatSth", "MLonSth", "MLTSth"] ] = rbspSatBDF[ ["MLatNth", "MLonNth", "MLTNth",\
             "MLatSth", "MLonSth", "MLTSth"] ].astype(float)
print rbspSatBDF.head()
####### DO THE SAME FOR SAT B #######
####### DO THE SAME FOR SAT B #######
```

```python
# We need to reformat both the files to represent the format
# that is used by DaViT (rbspFp.dat file stored in /sd-data/rbsp)
# The format is - yyyy mon day hour min glat_nth glon_nth glat_sth glon_sth
# First convert to datetime to reformat data
def convert_to_datetime(row):
        # Given a datestr and a time string convert to a python datetime obj.
        import datetime
        datecolName="dateStr"
        timeColName="timeStr"
        currDateStr = str( int( row[datecolName] ) )
    #     return currDateStr
        if row[timeColName] < 10:
            currTimeStr = "000" + str( int( row[timeColName] ) )
        elif row[timeColName] < 100:
            currTimeStr = "00" + str( int( row[timeColName] ) )
        elif row[timeColName] < 1000:
            currTimeStr = "0" + str( int( row[timeColName] ) )
        else:
            currTimeStr = str( int( row[timeColName] ) )
        return datetime.datetime.strptime( currDateStr\
                        + ":" + currTimeStr, "%Y%m%d:%H%M" )
    
    
rbspSatADF["date"] = rbspSatADF.apply( convert_to_datetime, axis=1 )
rbspSatBDF["date"] = rbspSatBDF.apply( convert_to_datetime, axis=1 )
```

```python
# SAT-A
rbspSatADF["year"] = rbspSatADF["date"].apply(lambda x: x.strftime('%Y'))
rbspSatADF["mon"] = rbspSatADF["date"].apply(lambda x: x.strftime('%-m'))
rbspSatADF["day"] = rbspSatADF["date"].apply(lambda x: x.strftime('%-d'))
rbspSatADF["hour"] = rbspSatADF["date"].apply(lambda x: x.strftime('%-H'))
rbspSatADF["min"] = rbspSatADF["date"].apply(lambda x: x.strftime('%-M'))
rbspSatADF["satName"] = rbspSatADF["sat"].apply(lambda x: x[-1])
# Only choose the selected cols
# rbspSatADF = rbspSatADF[ ["year", "mon", "day", "hour", "min",\
#                           "satName", "MLatNth", "MLonNth",\
#                           "date", "MLatSth", "MLonSth"] ]
# SAT-B
rbspSatBDF["year"] = rbspSatBDF["date"].apply(lambda x: x.strftime('%Y'))
rbspSatBDF["mon"] = rbspSatBDF["date"].apply(lambda x: x.strftime('%-m'))
rbspSatBDF["day"] = rbspSatBDF["date"].apply(lambda x: x.strftime('%-d'))
rbspSatBDF["hour"] = rbspSatBDF["date"].apply(lambda x: x.strftime('%-H'))
rbspSatBDF["min"] = rbspSatBDF["date"].apply(lambda x: x.strftime('%-M'))
rbspSatBDF["satName"] = rbspSatBDF["sat"].apply(lambda x: x[-1])
# rbspSatBDF = rbspSatBDF[ ["year", "mon", "day", "hour", "min",\
#                           "satName", "MLatNth", "MLonNth",\
#                           "date", "MLatSth", "MLonSth"] ]
print rbspSatADF.head()
print rbspSatBDF.head()
```

```python
# Concat the two DFs and sort them by time
rbspFpDF = pandas.concat( [rbspSatADF, rbspSatBDF],\
                         ignore_index=True )
rbspFpDF = rbspFpDF.sort('date').reset_index(drop=True)
rbspFpDF = rbspFpDF[ ["year", "mon", "day", "hour", "min",\
                          "satName", "MLatNth", "MLonNth",\
                          "MLatSth", "MLonSth"] ]
rbspFpDF.head()
```

```python
# Save every 5th row to keep things similar to the old formatsa
rbspFpDF.to_csv("../data/rbspFpMagn.csv", sep='\t', index=False)
```
