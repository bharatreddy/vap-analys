```python
import pandas
import datetime
```

```python
# read data from the lshell file
rbspSDVelDF = pandas.read_csv("../data/final_results_test.txt", sep=' ',\
                             infer_datetime_format=True, parse_dates=["eventTime"])
rbspSDVelDF.head()
```

```python
# If there are multiple measurements for
# a satellite at a given time then choose the
# one with min error in velocity Azim
rbspSDVelDF["azimPercError"] = rbspSDVelDF["errVelAzim"]/rbspSDVelDF["estVelAzim"]
rbspSDVelDF["magnPercError"] = rbspSDVelDF["errVelMagn"]/rbspSDVelDF["estVelMagn"]
minErrFitVels = rbspSDVelDF.groupby( ["sat", "eventTime"] )["azimPercError"].min().reset_index()
# Now join min error velocities with magnitudes...
rbspSDVelDF = pandas.merge( rbspSDVelDF, minErrFitVels,\
                           on=["sat", "eventTime", "azimPercError"] )
rbspSDVelDF.head(10)
```
